# r (read mode)
Read-only access – You cannot modify the file.

File must exist – If the file doesn’t exist, Python raises a FileNotFoundError.

File pointer starts at the beginning of the file.

Does not modify the file in any way.

```python
f1 = open('file_2.txt', 'r')
data = f1.read()
print(data)
```

### read only 10 characters

```python
f1 = open('file_2.txt', 'r')
print(f1.read(10))
```

### reading line by line of a file
* The readlines() method reads all the lines of a file and returns them as a list of strings, where each item is one line (including the newline character \n at the end)

```python
f1 = open('file_2.txt', 'r')
lines = f1.readlines()
for line in lines:
    print(line, end='')
f1.close()
```

# w - (Write Mode)
Creates a new file if it doesn’t exist.

Overwrites the file if it already exists (⚠ Data loss warning).

Only allows writing (reading is not possible unless reopened in read mode).

Returns a file object that can be used for writing.


```python
f2 = open('file_1.txt', 'w')
f2.write('Sai Varshath')
```

###      Writelines()
* The writelines() method in Python is used to write a list of strings to a file. 
* list is a list of strings, and each string ends with a newline character \n, so each name appears on a new line.
* f1.writelines(list) writes all the strings in the list to the file as they are.

```python
f1 = open('file_1.txt', 'w')
list = ['Venkateswarlu\n', 'Vasantha\n', 'Akhil\n', 'Sai ', 'Varshath']
f1.writelines(list)

```

# r+ mode(read and write)
* This mode is usefull for first read and then write
  
* Allows both reading and writing to the same file.

* Does NOT create a new file if it doesn’t exist (unlike "w+").

* Overwrites from the beginning (doesn’t erase the entire file).

* File pointer starts at the beginning of the file.



```python
f2 = open('file_1.txt', 'r+')
print(f2.read())
f2.write('Konduri Akhil')
```

```python
f2 = open('file_1.txt', 'r+')
f2.write('Konduri ')
print(f2.read())
f2.seek(0)
print(f2.read())

```

tell() - To know the pointer position 


```python
f2 = open('file_1.txt', 'r+')
print(f2.tell())
f2.write('Konduri ')
print(f2.tell())
f2.seek(0)
print(f2.tell())
print(f2.read())
print(f2.tell())
```

# w+ mode (read and write )

1. **Creates a New File If It Doesn’t Exist**  
   - If the file doesn’t exist, `w+` will create a new file automatically.  

2. **Erases Existing Content**  
   - If the file **already exists**, `w+` **clears all its content** before writing.  
   - This is different from `r+`, which **preserves existing content** unless explicitly overwritten.  

3. **Allows Both Reading and Writing**  
   - You can write and then read or read and then write.  

4. **Pointer Starts at the Beginning**  
   - But since `w+` **erases content**, reading after opening won’t return old data.  


```python
f1 = open('file_11.txt', 'w+')
print(f1.tell())
f1.write('This is Konduri')
print(f1.tell())
print(f1.read())

```

```python
f1 = open('file_1.txt', 'w+')
print(f1.tell())
f1.write('This is Konduri')
print(f1.tell())
f1.seek(0)
print(f1.tell())
print(f1.read())
f1.close()

```

# a (append mode)
   - Opens the file for writing.  
   - **Pointer starts at the end** of the file.  
   - **Does NOT allow reading**.  
   - **Creates a new file** if it doesn’t exist.  
   - **All writes happen at the end** (cannot overwrite existing content). 

```python
f1 = open('file_1.txt', 'a')
print(f1.tell())
f1.write('Akhil')
print(f1.tell())
print('Is Writable: ', f1.writable())
print('Is Readable: ', f1.readable())
```

```python
f1 = open('file_1.txt', 'a')
print(f1.tell())
f1.write(' Akhil')
print(f1.tell())
```

# a+ mode (read and append)
   - Opens the file for both **reading and appending**.  
   - **Pointer starts at the end** when writing, but you can read from the beginning.  
   - **Creates a new file** if it doesn’t exist.  
   - **All writes always happen at the end**, even if you move the pointer.

```python
f1 = open('file_2.txt', 'a+')
f1.write('This is python course')
print(f1.tell())
f1.seek(0)
print(f1.tell())
print(f1.read())
print(f1.tell())
f1.seek(0)
f1.write(', My name is Konduri Akhil')
print(f1.tell())
f1.seek(0)
print(f1.read())


```

# With statement
The with statement is used to open and automatically close resources — like files — without having to call close() manually.

🧠 Why is it useful?

* Prevents file from being left open

* Cleaner and shorter code

* Avoids resource leaks

* Works with other resources too (like sockets, database connections, etc.)

```python
with open('file_1.txt', 'r') as f:
    name1 = f.readline()
    print(name1, end='')
    name2 = f.readline()
    print(name2, end='')
    print('Is file closed: ', f.closed)

print('Is file closed: ', f.closed)
```
